```python
#importing numpy and pandas
import numpy as np
import pandas as pd

#importing for visualization
import matplotlib.pyplot as plt

%matplotlib inline

r_s=123456
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Epileptic Seizure Recognition.csv")
df
```

```python
df.isnull().values.any()
```

```python
#target variable
y=df['y']
```

```python
y=df['y']
y = y.apply(lambda x: 1 if x == 1 else 0)
y.unique()
```

```python
X=df.drop('y',axis=1)
X.head()
```

```python
import pandas as pd

# Assuming you have a DataFrame X, you can drop non-numeric columns like this:
numeric_columns = X.select_dtypes(include=['number'])  # Select numeric columns
X_numeric = X[numeric_columns.columns]  # Create a DataFrame with only numeric columns
from sklearn.preprocessing import StandardScaler

# Scale the numeric data
scale = StandardScaler().fit(X_numeric)
x_scaled = scale.transform(X_numeric)
x_scaled
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)  #two component
principalComponents = pca.fit_transform(x_scaled)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
principalDf.head()
```

```python
#Concating Principal components and y varaible
PCAdf= pd.concat([principalDf, y], axis = 1)
PCAdf.head()
```

T-sne is an another dimensionality reduction techinique

```python
from sklearn.manifold import TSNE

TNSEdf = TSNE(random_state=r_s).fit_transform(x_scaled)
TNSEdf.shape
tnsedf = pd.DataFrame(TNSEdf)
tnsedf['y']=y
tnsedf.head()
```

*K-means*

```python
#K-Means
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=5)
```

Implementing kmeans in pca

```python
#implementing kmeans in pcadf
kmeansPCA = kmeans.fit(principalDf)
#Getting the cluster labels
labels = kmeans.predict(principalDf)
PK_df= PCAdf
PK_df['kclust']=labels
PK_df['kclust'].unique()
```

Implementing kmeans in T-SNE

```python
#K-Means
from sklearn.cluster import KMeans
# Number of clusters
kmeans = KMeans(n_clusters=5)
#implementing kmeans in tnsedf
kmeanstnse = kmeans.fit(TNSEdf)
#Getting the cluster labels
k_labels = kmeans.predict(TNSEdf)
```

```python
TK_df = tnsedf
TK_df['cluster']=k_labels
TK_df.info()
```

```python
TK_df['cluster'].unique()
```

Splitting the data into train and test

```python
#splitting into x and y varialbles

pk_x=PK_df.loc[:,['principal component 1','principal component 2','kclust']]
pk_y=PK_df['y']

tk_x=TK_df.loc[:,[0,1,'cluster']]
tk_y=TK_df['y']
```

```python
from sklearn.model_selection import train_test_split

#PCA +k-means
pkx_train, pkx_test, pky_train, pky_test = train_test_split(pk_x,pk_y,
test_size = 0.2,random_state = 101)

#T-SNE + k-means
tkx_train, tkx_test, tky_train, tky_test = train_test_split(tk_x,tk_y,
test_size = 0.2, random_state = 101)
```

metrix of classfication

```python
#Confusion matrix and accuraccy score
from sklearn.metrics import confusion_matrix, accuracy_score
#Cross validation
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

Logistic regression

```python
from sklearn.linear_model import LogisticRegression

# Convert feature names to strings
tkx_train.columns = tkx_train.columns.astype(str)
tkx_test.columns = tkx_test.columns.astype(str)

# Create a logistic regression model
logmodel = LogisticRegression()

# Fit the logistic regression model to the training data
logmodel.fit(tkx_train, tky_train)

# Predict using the trained model
tklogpred = logmodel.predict(tkx_test)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()

logmodel.fit(pkx_train,pky_train)
pklogpred = logmodel.predict(pkx_test)


logmodel.fit(tkx_train,tky_train)
tklogpred = logmodel.predict(tkx_test)
```

```python
print(confusion_matrix(pky_test, pklogpred))

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(pky_test, pklogpred)
precision = precision_score(pky_test, pklogpred)
recall = recall_score(pky_test, pklogpred)
f1 = f1_score(pky_test, pklogpred)

print(f"Accuracy: {round(accuracy, 2) * 100}%")
print(f"Precision: {round(precision, 2) * 100}%")
print(f"Recall: {round(recall, 2) * 100}%")
print(f"F1-score: {round(f1, 2) * 100}%")
```

```python
print(confusion_matrix(tky_test, tklogpred))

accuracy = accuracy_score(tky_test, tklogpred)
precision = precision_score(tky_test, tklogpred)
recall = recall_score(tky_test, tklogpred)
f1 = f1_score(tky_test, tklogpred)

print(f"Accuracy: {round(accuracy, 2) * 100}%")
print(f"Precision: {round(precision, 2) * 100}%")
print(f"Recall: {round(recall, 2) * 100}%")
print(f"F1-score: {round(f1, 2) * 100}%")
```

```python
#cross validation
LOGCV = (cross_val_score(logmodel, tkx_train, tky_train, cv=k_fold, n_jobs=1, scoring = 'accuracy').mean())
```
